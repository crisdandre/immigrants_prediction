### **Turistas Anuales**

Este notebook hace un simple análisis de la cantidad Turistas extranjeros que llegaron a España en Avión por año.

Datos disponibles en: https://es.statista.com/estadisticas/474883/llegadas-de-turistas-extranjeros-a-espana-en-avion/

```python
# Dependencias básicas
import pandas as pd
import numpy as np

# Dependencias de visualización
from matplotlib import pyplot as plt
import seaborn as sns
import altair as alt

# Dependencias de machine learning
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
# Lectura de dataframe
df = pd.read_csv("../7 -Turistas Anuales/datos_turistas.csv")
```

```python
# Visualización de dataframe
df
```

```python
df.rename(columns={"Año": "Year", "Millones de turistas": "Total"}, inplace = True)
```

```python
# Eliminamos asteriscos
df.Year = df.Year.str.strip(to_strip="*")
```

```python
df.Total = df.Total.apply(lambda x: x * 1000000)
df.Total = df.Total.astype("int")
```

```python
df
```

```python
plt.figure(figsize=(9, 4))
plt.plot(df.Year, df.Total)
plt.xticks(rotation=45)
plt.xlabel("Año")
plt.ylabel("Millones de turistas")
plt.show()
```

```python
# Renombrar colomna Total
df.rename({'Total' : 'Number of Turist'}, inplace = True, axis = 1)

# Exportación para trabajo en modelo
df.to_csv("../13 - Exports (preprocesamiento)/turismo.csv", index = False)
```

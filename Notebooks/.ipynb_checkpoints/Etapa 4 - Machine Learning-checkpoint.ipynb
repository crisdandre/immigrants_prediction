### *Preparar Datos*

```python
# Importar el dataset
df = pd.read_csv("../13 - Exports (preprocesamiento)/inmigrantes_merge.csv")

df.info()
df
```

Antes de proceder, debemos resolver la variable "Year" a una variable ordinal (los regímenes políticos ya están en formato ordinal) y el resto de variables categóricas a variables dummy.

En el caso de Year, simplemente restaremos 2007 a la columna entera, y para el resto de las variables objeto usaremos la funcion *.get_dummies()*.

```python
# hacer copia del df
df_copy = df.copy()

# Transformar Year a variable ordinal de 1 (2008) a 15 (2022)
df_copy['Year'] = df_copy['Year'] - 2007

# Generar variables dummies a partir de nuestras variables categóricas "object" (no ordinales)
df_copy = pd.get_dummies(df_copy)

# Convertir las variables dummies booleanas en "int"
col_bool = df_copy.select_dtypes(include = ['bool']).columns
df_copy[col_bool] = df_copy[col_bool].astype(int)

# Verificar cambio
df_copy.info()
df_copy
```

### *Separar conjunto de entrenamiento y prueba*

```python
# Separar variables input y variable target "Immigrant count"
X = df_copy.drop("Immigrant count", axis = 1) # variables predictoras
y = df_copy["Immigrant count"]  # Target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 58) # separar datos en conjunto train y test en un 75% / 25%
scaler = RobustScaler() # definir scaler de datos 
X_train = scaler.fit_transform(X_train) # escalar los datos de entrenamiento 
X_test = scaler.fit_transform(X_test) # # escalar los datos de prueba
```

### *Modelos de Machine Learning*

#### Red neuronal recurrente de corto-largo plazo

```python
# Function to create rolling windows
def create_rolling_windows(df, window_size):
    X = []
    Y = []
    
    for i in range(len(df) - window_size):
        X.append(df.iloc[i:i + window_size].values)
        Y.append(df.iloc[i + window_size].values)
    
    X = np.array(X)
    Y = np.array(Y)
    return X, Y

# Step 1: Create rolling windows
window_size = 5
X, Y = create_rolling_windows(df_copy[['Immigrant count']], window_size)

# Step 2: Reshape the data for RNN input
# X should be of shape (samples, time_steps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))  # 1 feature ('Value')

# Step 3: Apply MinMaxScaler to scale the data between 0 and 1
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_Y = MinMaxScaler(feature_range=(0, 1))

# Reshape X and Y for scaling
X_reshaped = X.reshape(-1, 1)  # Flatten X for scaling
Y_reshaped = Y.reshape(-1, 1)  # Flatten Y for scaling

# Fit the scaler on the reshaped data and transform
X_scaled = scaler_X.fit_transform(X_reshaped).reshape(X.shape)  # Reshape back to original shape
Y_scaled = scaler_Y.fit_transform(Y_reshaped)

# Step 4: Split into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y_scaled, test_size=0.2, shuffle=False)

# Step 5: Build the RNN/LSTM model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(window_size, 1)))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Step 6: Train the model
model.fit(X_train, Y_train, epochs=80, validation_data=(X_test, Y_test))

# Step 7: Make predictions
Y_pred_scaled = model.predict(X_test)

# Step 8: Inverse transform predictions and test data to original scale
Y_pred = scaler_Y.inverse_transform(Y_pred_scaled)
Y_test_original = scaler_Y.inverse_transform(Y_test)

# Display the model summary
model.summary()
```

La red neuronal recurrente de memoria a corto-largo plazo (LSTM) anterior sigue una estructura sencilla sobre la predicción a la que intenta llegar. La capa de densidad sigue una lógica de cálculo como sigue: 50 unidades de predicción por cada unidad de tiempo de la secuencia de input; luego los parámetros entrenables suman 10.400 al ser el resultado de la multiplicación:

 4 * ((núm. de unidades * (núm. de características de input + 1)) + núm. de unidades^2)

Esto incluye las cuatro variables necesarias para la red neuronal (input, output, forget, y estado de célula)

Para la parte de la capa, obtenemos un valor de resultado (shape None, 1) y tenemos 51 parámetros de cálculo como lo define la fórmula siguiente:

Núm. de unidades (50 de LSTM) * Número de outputs (1) + un grado de bias (1)

Los parámetros suman 10,451 mismos que se actualizan conforme se actualice el modelo.
Tenemos 20.904 parámetros de optimización. Se basan en el cálculo del optimizador Adam y variables internas.

```python
# Graficamos el número de parámetros por cada layer en el modelo

layers = ['LSTM Layer', 'Dense Layer']
params = [10400, 51]

plt.bar(layers, params, color=['skyblue', 'lightgreen'])
plt.title('Trainable Parameters by Layer')
plt.ylabel('Number of Parameters')
plt.show()
```

```python
# Definimos diccionario para acceder a model.history
history = model.history

# Graficamos desempeño de la pérdida en train y validation
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

En este gráfico, observamos la frecuencia con la que se pierde la certeza en los sets de validación y de entrenamiento. Con entrenamiento tenemos un desempeño estable, mientras que la validación mantiene unas milésimas de error por encima de entrenamiento. También nos sugiere que el entrenamiento se terminó en la epoch 6, al no tener resultados más relevante en las epochs siguientes. Lo comprobamos con el número de keys en el diccionaro history:
